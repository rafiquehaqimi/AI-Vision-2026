```python
# ==========================================================
# FIX ME CHALLENGE: PROFIL PERMULAAN ROBOT
# ==========================================================

# Menggantikan nilai dummy dengan data sebenar/contoh yang betul
robot_name = "KV-TRON Safety Patrol"
robot_id = 205
battery_level = 95
operation_mode = "Pemeriksaan Bengkel"
robot_location = "Bengkel TVET A"

# FIX ME: Menggunakan fungsi input() untuk menerima nama jurutera secara dinamik
engineer_name = input("Masukkan nama jurutera: ")

print("\n====================================")
print("        PROFIL PERMULAAN ROBOT")
print("====================================")
print("Nama Robot    :", robot_name)
print("ID Robot      :", robot_id)
print("Bateri        :", battery_level, "%")
print("Mod Operasi   :", operation_mode)
print("Lokasi        :", robot_location)
print("Jurutera      :", engineer_name)

# Logik semakan status kelayakan profil
lengkap = (
    robot_name != "FIX_ME"
    and robot_id is not None
    and battery_level is not None
    and operation_mode != "FIX_ME"
    and robot_location != "FIX_ME"
    and engineer_name != "FIX_ME"
    and engineer_name.strip() != "" # Tambahan: memastikan input tidak dibiarkan kosong
)

print("------------------------------------")
print("[ROBOT ONLINE] Identiti lengkap." if lengkap
      else "[BELUM LENGKAP] Baiki semua bahagian FIX ME.")
```

#Latihan Praktikal: Sistem Pendaftaran Ahli Kelab Robotik
Situasi

Anda merupakan seorang pembantu makmal AI di Kolej Vokasional. Hari ini anda diminta membina satu program ringkas untuk mendaftarkan ahli baharu Kelab Robotik.

Program anda perlu mendapatkan maklumat daripada pengguna dan memaparkannya semula dengan kemas.

Tugasan

Bina satu program yang boleh:

Meminta pengguna memasukkan:
Nama

*   Umur
*   Program Pengajian

*   Semester
*   Nama Robot Pilihan

Paparkan semua maklumat tersebut dengan format yang kemas.
Paparkan jenis data (data type) bagi setiap maklumat yang dimasukkan.

image.png

#Cabaran

Ubah program supaya pengguna turut memasukkan:

*   Tahun Pengajian
*   Minat dalam AI

Kemudian paparkan maklumat tersebut.


```python
# 1. Mengambil input daripada pengguna (Termasuk bahagian Cabaran)
nama = input("Masukkan Nama: ")
umur = input("Masukkan Umur: ")
program = input("Masukkan Program Pengajian: ")
semester = input("Masukkan Semester: ")
robot_pilihan = input("Masukkan Nama Robot Pilihan: ")
tahun_pengajian = input("Masukkan Tahun Pengajian (Cabaran): ")
minat_ai = input("Adakah anda berminat dalam AI? (Cabaran): ")

# 2. Memaparkan maklumat pendaftaran
print("\n========================================")
print("     PENDAFTARAN AHLI KELAB ROBOTIK     ")
print("========================================")
print(f"Nama            : {nama}")
print(f"Umur            : {umur}")
print(f"Program         : {program}")
print(f"Semester        : {semester}")
print(f"Robot Pilihan   : {robot_pilihan}")
print(f"Tahun Pengajian : {tahun_pengajian}")
print(f"Minat dalam AI  : {minat_ai}")

# 3. Memaparkan jenis data (data type) bagi setiap maklumat
print("\n========================================")
print("Jenis Data")
print("========================================")
print(f"Nama            : {type(nama)}")
print(f"Umur            : {type(umur)}")
print(f"Program         : {type(program)}")
print(f"Semester        : {type(semester)}")
print(f"Robot Pilihan   : {type(robot_pilihan)}")
print(f"Tahun Pengajian : {type(tahun_pengajian)}")
print(f"Minat dalam AI  : {type(minat_ai)}")
```
